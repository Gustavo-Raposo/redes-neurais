```python
import pandas as pd
```

```python
# Path to the txt file
file_path = './SOBRADINHO_DIARIA_1931_2016.txt'

# Reading file
with open(file_path, 'r') as file:
    lines = file.readlines()

# Creating a DataFrame
df = pd.DataFrame(lines, columns=['Flow'])

# Strip newline characters from the end of each line
df['Flow'] = df['Flow'].str.strip()

print(df)
```

```python
#função para fazer o deslocamentos das saídas e entradas
def deslocamento(entradas,saidas,tabela):
    #cria uma cópia da tabela
    tabela1=tabela.copy()
    #desloca a coluna flow i vezes
    for i in range(1,entradas+1):
        tabela1.insert(loc=0, column='Entrada'+str((entradas-i)+1), value=tabela['Flow'].shift(i))
    #desloca a coluna 
    for i in range(1,saidas):
        tabela1['Flow'+ str(i)]=tabela['Flow'].shift(-1*i)
    return tabela1
    
```
